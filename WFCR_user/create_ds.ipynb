```python
from wfcr import WFCR
import os

raw_photos = "dataset/original"
img_det = "dataset/detected"
bbox_path = "dataset/bbox"
mask_path = "dataset/mask"

reader = WFCR()
for num, file in enumerate(os.listdir(raw_photos)):
    if not num == 1500:
        print(f"{num+1}. {file} processing")
        file_path = os.path.join(raw_photos, file)
        img, img_bbox, img_mask = reader.detect(file_path)

        img.save(os.path.join(img_det, file))    
        img_bbox.save(os.path.join(bbox_path, file))
        img_mask.save(os.path.join(mask_path, file))
    else:
        break
```
